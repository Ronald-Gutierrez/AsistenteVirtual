# Simulaci√≥n de Interacci√≥n para Luc√≠a en MiBolsillo

La siguiente simulaci√≥n representa la interacci√≥n de Luc√≠a con la plataforma "MiBolsillo", 
utilizando visualizaciones y respuestas basadas en datos simulados.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos desde el CSV proporcionado
df = pd.read_csv("/Users/johar/Desktop/hackathon/lucia_data.csv")
```

```python
def plot_expenses(df):
    """Genera un gr√°fico de barras con desgloses de categor√≠as."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categor√≠a")["Monto"].sum()
    sorted_expenses = grouped_expenses.sort_values()
    
    plt.figure(figsize=(10, 6))
    sorted_expenses.plot(kind='barh', color='skyblue')
    plt.xlabel('Monto (S/)')
    plt.title('Gastos del √∫ltimo mes por categor√≠a')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def meals_out_expense(df):
    """Devuelve una respuesta sobre el gasto en comidas fuera."""
    meals_out_amount = df[df["Categor√≠a"] == "Comidas Fuera"]["Monto"].sum()
    previous_month_amount = df[df["Categor√≠a"] == "Comidas Fuera"][:-10]["Monto"].sum()  
    increase_percentage = ((meals_out_amount - previous_month_amount) / abs(previous_month_amount)) * 100
    response = f"Gastaste S/ {-meals_out_amount} en comidas fuera, lo que representa un {increase_percentage:.2f}% m√°s que el mes anterior."
    return response

def recommendations(df):
    """Devuelve recomendaciones para reducir gastos."""
    potential_savings = df[df["Categor√≠a"] == "Comidas Fuera"]["Monto"].sum() * 0.10
    response = f"Si decides reducir tus comidas fuera en solo un 10%, podr√≠as ahorrar S/ {potential_savings} al mes o S/ {potential_savings*12} al a√±o."
    return response
```

```python
print("Plataforma: ¬°Hola, Luc√≠a! Bienvenida de nuevo a MiBolsillo.\n")
print("üìä A continuaci√≥n, te presentamos un resumen visual de tus gastos del √∫ltimo mes.\n")
plot_expenses(df)
```

```python
print("Luc√≠a: ¬øCu√°nto gast√© en comidas fuera el √∫ltimo mes?")
print(f"Plataforma: {meals_out_expense(df)}")
```

```python
print("Luc√≠a: ¬°Vaya! ¬øTienes alguna recomendaci√≥n para reducir eso?")
print(f"Plataforma: {recommendations(df)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de Luc√≠a desde un CSV
df = pd.read_csv("/Users/johar/Desktop/hackathon/lucia_data.csv")

# Funciones de ayuda

def plot_expenses(df):
    """Genera un gr√°fico de barras con desgloses de categor√≠as."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categor√≠a")["Monto"].sum()
    sorted_expenses = grouped_expenses.sort_values()
    
    plt.figure(figsize=(6, 2))
    sorted_expenses.plot(kind='barh', color='skyblue')
    plt.xlabel('Monto (S/)')
    plt.title('Gastos del √∫ltimo mes por categor√≠a')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def highlight_expense(df, category):
    """Resalta una categor√≠a espec√≠fica en el gr√°fico."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categor√≠a")["Monto"].sum()
    colors = ['skyblue' if cat != category else 'red' for cat in grouped_expenses.index]
    
    plt.figure(figsize=(6, 2))
    grouped_expenses.plot(kind='barh', color=colors)
    plt.xlabel('Monto (S/)')
    plt.title(f'Gastos del √∫ltimo mes con {category} resaltado')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def savings_projection(current_expense, reduction_percentage):
    """Muestra una gr√°fica de proyecci√≥n de ahorros en funci√≥n del tiempo."""
    months = list(range(1, 13))
    savings = [current_expense * reduction_percentage * month for month in months]
    
    plt.figure(figsize=(6, 2))
    plt.plot(months, savings, marker='o')
    plt.xlabel('Meses')
    plt.ylabel('Ahorro acumulado (S/)')
    plt.title('Proyecci√≥n de ahorros')
    plt.grid(True, which='both', linestyle='--', alpha=0.7)
    plt.show()

# Simulaci√≥n de la conversaci√≥n

print("Usuario: Luc√≠a")
print("\nAcci√≥n: Luc√≠a ingresa a la plataforma \"MiBolsillo\" despu√©s de haber conectado previamente su perfil del BCP.")
print("\nPlataforma: ¬°Hola, Luc√≠a! Bienvenida de nuevo a MiBolsillo.")
print("\nüìä A continuaci√≥n, te presentamos un resumen visual de tus gastos del √∫ltimo mes.\n")
plot_expenses(df)

print("\nLuc√≠a: ¬øCu√°nto gast√© en comidas fuera el √∫ltimo mes?")
print("Plataforma: Gastaste S/ 850 en comidas fuera, lo que representa un 20% m√°s que el mes anterior.")
print("üìâ Aqu√≠ puedes ver c√≥mo se compara con tus otros gastos.\n")
highlight_expense(df, "Comidas Fuera")

print("\nLuc√≠a: ¬°Vaya! ¬øTienes alguna recomendaci√≥n para reducir eso?")
print("Plataforma: Claro, Luc√≠a. Si decides reducir tus comidas fuera en solo un 10%, podr√≠as ahorrar S/ 85 al mes o S/ 1020 al a√±o.")
print("üìà Aqu√≠ hay una proyecci√≥n de c√≥mo se ver√≠an tus ahorros si implementas este cambio.\n")
savings_projection(850, 0.10)

print("\nLuc√≠a: Eso suena bien. ¬øQu√© otros gastos puedo reducir?")
print("Plataforma: Bas√°ndonos en tu perfil, tambi√©n priorizas las compras de ropa. Algunas opciones podr√≠an incluir:")
print("\n- Buscar ofertas o descuentos en tus tiendas favoritas.")
print("- Establecer un presupuesto mensual para compras de ropa.")
print("- Considerar ventas de segunda mano o intercambios con amigos.")
print("\nüìå ¬øTe gustar√≠a ver una simulaci√≥n de cu√°nto podr√≠as ahorrar si reduces tus gastos en ropa en un 15%?")
print("\nLuc√≠a: S√≠, me gustar√≠a verlo.")
print("Plataforma: Perfecto.")
print("üìä Aqu√≠ tienes una proyecci√≥n visual de tus ahorros si decides implementar este cambio en tus gastos de ropa.\n")
savings_projection(df[df["Categor√≠a"] == "Ropa"]["Monto"].sum(), 0.15)

print("\nLuc√≠a: ¬°Eso es √∫til! Tambi√©n he estado pensando en comprar un auto. ¬øQu√© necesitar√≠a hacer para lograrlo?")
# Puedes adaptar el valor "X" y "Y" a tus necesidades.
print("Plataforma: ¬°Excelente meta, Luc√≠a!")
print("üöó Bas√°ndonos en el precio promedio de un auto que podr√≠a interesarte, necesitar√≠as ahorrar aproximadamente S/ X al mes durante Y a√±os.")
print("üìà Aqu√≠ tienes un roadmap visual que muestra los pasos y el progreso hacia esa meta.\n")
# Aqu√≠ puedes agregar la funci√≥n para mostrar el roadmap para la compra del auto.

# NOTA: Es necesario que adaptes algunas partes, como la ruta del archivo CSV y los c√°lculos espec√≠ficos, para que se ajusten a la realidad de tus datos y necesidades.
```
