# Simulación de Interacción para Lucía en MiBolsillo

La siguiente simulación representa la interacción de Lucía con la plataforma "MiBolsillo", 
utilizando visualizaciones y respuestas basadas en datos simulados.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos desde el CSV proporcionado
df = pd.read_csv("/Users/johar/Desktop/hackathon/lucia_data.csv")
```

```python
def plot_expenses(df):
    """Genera un gráfico de barras con desgloses de categorías."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categoría")["Monto"].sum()
    sorted_expenses = grouped_expenses.sort_values()
    
    plt.figure(figsize=(10, 6))
    sorted_expenses.plot(kind='barh', color='skyblue')
    plt.xlabel('Monto (S/)')
    plt.title('Gastos del último mes por categoría')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def meals_out_expense(df):
    """Devuelve una respuesta sobre el gasto en comidas fuera."""
    meals_out_amount = df[df["Categoría"] == "Comidas Fuera"]["Monto"].sum()
    previous_month_amount = df[df["Categoría"] == "Comidas Fuera"][:-10]["Monto"].sum()  
    increase_percentage = ((meals_out_amount - previous_month_amount) / abs(previous_month_amount)) * 100
    response = f"Gastaste S/ {-meals_out_amount} en comidas fuera, lo que representa un {increase_percentage:.2f}% más que el mes anterior."
    return response

def recommendations(df):
    """Devuelve recomendaciones para reducir gastos."""
    potential_savings = df[df["Categoría"] == "Comidas Fuera"]["Monto"].sum() * 0.10
    response = f"Si decides reducir tus comidas fuera en solo un 10%, podrías ahorrar S/ {potential_savings} al mes o S/ {potential_savings*12} al año."
    return response
```

```python
print("Plataforma: ¡Hola, Lucía! Bienvenida de nuevo a MiBolsillo.\n")
print("📊 A continuación, te presentamos un resumen visual de tus gastos del último mes.\n")
plot_expenses(df)
```

```python
print("Lucía: ¿Cuánto gasté en comidas fuera el último mes?")
print(f"Plataforma: {meals_out_expense(df)}")
```

```python
print("Lucía: ¡Vaya! ¿Tienes alguna recomendación para reducir eso?")
print(f"Plataforma: {recommendations(df)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de Lucía desde un CSV
df = pd.read_csv("/Users/johar/Desktop/hackathon/lucia_data.csv")

# Funciones de ayuda

def plot_expenses(df):
    """Genera un gráfico de barras con desgloses de categorías."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categoría")["Monto"].sum()
    sorted_expenses = grouped_expenses.sort_values()
    
    plt.figure(figsize=(6, 2))
    sorted_expenses.plot(kind='barh', color='skyblue')
    plt.xlabel('Monto (S/)')
    plt.title('Gastos del último mes por categoría')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def highlight_expense(df, category):
    """Resalta una categoría específica en el gráfico."""
    grouped_expenses = df[df["Tipo"] == "Gasto"].groupby("Categoría")["Monto"].sum()
    colors = ['skyblue' if cat != category else 'red' for cat in grouped_expenses.index]
    
    plt.figure(figsize=(6, 2))
    grouped_expenses.plot(kind='barh', color=colors)
    plt.xlabel('Monto (S/)')
    plt.title(f'Gastos del último mes con {category} resaltado')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

def savings_projection(current_expense, reduction_percentage):
    """Muestra una gráfica de proyección de ahorros en función del tiempo."""
    months = list(range(1, 13))
    savings = [current_expense * reduction_percentage * month for month in months]
    
    plt.figure(figsize=(6, 2))
    plt.plot(months, savings, marker='o')
    plt.xlabel('Meses')
    plt.ylabel('Ahorro acumulado (S/)')
    plt.title('Proyección de ahorros')
    plt.grid(True, which='both', linestyle='--', alpha=0.7)
    plt.show()

# Simulación de la conversación

print("Usuario: Lucía")
print("\nAcción: Lucía ingresa a la plataforma \"MiBolsillo\" después de haber conectado previamente su perfil del BCP.")
print("\nPlataforma: ¡Hola, Lucía! Bienvenida de nuevo a MiBolsillo.")
print("\n📊 A continuación, te presentamos un resumen visual de tus gastos del último mes.\n")
plot_expenses(df)

print("\nLucía: ¿Cuánto gasté en comidas fuera el último mes?")
print("Plataforma: Gastaste S/ 850 en comidas fuera, lo que representa un 20% más que el mes anterior.")
print("📉 Aquí puedes ver cómo se compara con tus otros gastos.\n")
highlight_expense(df, "Comidas Fuera")

print("\nLucía: ¡Vaya! ¿Tienes alguna recomendación para reducir eso?")
print("Plataforma: Claro, Lucía. Si decides reducir tus comidas fuera en solo un 10%, podrías ahorrar S/ 85 al mes o S/ 1020 al año.")
print("📈 Aquí hay una proyección de cómo se verían tus ahorros si implementas este cambio.\n")
savings_projection(850, 0.10)

print("\nLucía: Eso suena bien. ¿Qué otros gastos puedo reducir?")
print("Plataforma: Basándonos en tu perfil, también priorizas las compras de ropa. Algunas opciones podrían incluir:")
print("\n- Buscar ofertas o descuentos en tus tiendas favoritas.")
print("- Establecer un presupuesto mensual para compras de ropa.")
print("- Considerar ventas de segunda mano o intercambios con amigos.")
print("\n📌 ¿Te gustaría ver una simulación de cuánto podrías ahorrar si reduces tus gastos en ropa en un 15%?")
print("\nLucía: Sí, me gustaría verlo.")
print("Plataforma: Perfecto.")
print("📊 Aquí tienes una proyección visual de tus ahorros si decides implementar este cambio en tus gastos de ropa.\n")
savings_projection(df[df["Categoría"] == "Ropa"]["Monto"].sum(), 0.15)

print("\nLucía: ¡Eso es útil! También he estado pensando en comprar un auto. ¿Qué necesitaría hacer para lograrlo?")
# Puedes adaptar el valor "X" y "Y" a tus necesidades.
print("Plataforma: ¡Excelente meta, Lucía!")
print("🚗 Basándonos en el precio promedio de un auto que podría interesarte, necesitarías ahorrar aproximadamente S/ X al mes durante Y años.")
print("📈 Aquí tienes un roadmap visual que muestra los pasos y el progreso hacia esa meta.\n")
# Aquí puedes agregar la función para mostrar el roadmap para la compra del auto.

# NOTA: Es necesario que adaptes algunas partes, como la ruta del archivo CSV y los cálculos específicos, para que se ajusten a la realidad de tus datos y necesidades.
```
